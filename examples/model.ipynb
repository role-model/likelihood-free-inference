```python
from roler.model import sample_from_prior, run_simulation
from roler.types import ModelPrior
from roler.distributions import *
```

```python
prior = ModelPrior(
    individuals_local=IntegerDistribution(UniformDistribution(50, 150)),
    individuals_meta = IntegerDistribution(UniformDistribution(500, 1500)),
    species_meta = IntegerDistribution(UniformDistribution(20, 80)),
    speciation_local = UniformDistribution(0.01, 0.1),
    speciation_meta = UniformDistribution(0.01, 0.1),
    extinction_meta = UniformDistribution(0.01, 0.1),
    env_sigma = UniformDistribution(0.4, 0.6),
    trait_sigma = ConstantDistribution(1),
    comp_sigma = UniformDistribution(0.4, 0.6),
    dispersal_prob = UniformDistribution(0.05, 0.15),
    mutation_rate = UniformDistribution(0.005, 0.015),
    equilib_escape = ConstantDistribution(1),
    num_basepairs = IntegerDistribution(UniformDistribution(200, 300)),
    init_type = ConstantDistribution('oceanic_island'),
    niter = IntegerDistribution(UniformDistribution(5000, 15000)),
    niterTimestep = IntegerDistribution(UniformDistribution(5, 20))
)
```

```python
simple_prior = ModelPrior(
    individuals_local=IntegerDistribution(UniformDistribution(100, 100)),
    individuals_meta = IntegerDistribution(UniformDistribution(1000, 1000)),
    species_meta = IntegerDistribution(UniformDistribution(20, 80)),
    speciation_local = UniformDistribution(0.01, 0.95),
    speciation_meta = UniformDistribution(0.01, 0.95),
    extinction_meta = UniformDistribution(0.01, 0.95),
    env_sigma = UniformDistribution(0.5, 0.5),
    trait_sigma = ConstantDistribution(1),
    comp_sigma = UniformDistribution(0.5, 0.5),
    dispersal_prob = UniformDistribution(0.1, 0.1),
    mutation_rate = UniformDistribution(0.01, 0.01),
    equilib_escape = ConstantDistribution(1),
    num_basepairs = IntegerDistribution(UniformDistribution(250, 250)),
    init_type = ConstantDistribution('oceanic_island'),
    niter = IntegerDistribution(UniformDistribution(1000, 1000)),
    niterTimestep = IntegerDistribution(UniformDistribution(10, 10))
)
```

```python
original_param_distribution = {
    "individuals_local": 100,
    "individuals_meta": 1000,
    "species_meta": 50,
    "speciation_local": 0.05,
    "speciation_meta": 0.05,
    "extinction_meta": 0.05,
    "env_sigma": 0.5,
    "trait_sigma": 1,
    "comp_sigma": 0.5,
    "dispersal_prob": 0.1,
    "mutation_rate": 0.01,
    "equilib_escape": 1,
    "num_basepairs": 250,
    "init_type": 'oceanic_island',
    "niter": 10000,
    "niterTimestep": 10
}
```

```python
import matplotlib.pyplot as plt

stats_df = run_simulation(sample_from_prior(simple_prior))
plt.figure(figsize=(10, 6))
plt.plot(stats_df['iteration'], stats_df['rich'], color='blue')
plt.xlabel('Iteration')
plt.ylabel('Richness')
plt.title('Richness vs Iteration')
plt.show()
```
